```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Apply OLS
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
# we apply Gradient descent assuming slop(coef_)=78.35
# starting value for intercept b =0
y_pred=((78.35 * X)+100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green',label='OLS')
plt.plot(X,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
m= 78.35
b=100
loss_slope= -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Let's take lerning rate=0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
# Calculate new intercept
b=b-step_size
b

```

```python
y_pred1=((78.35 * X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='green',label='b={}'.format(b))
plt.plot(X,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
lr=0.1

epochs=100

for i in range(epochs):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    b=b-(lr*loss_slope)
    y_pred=m*X+b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
