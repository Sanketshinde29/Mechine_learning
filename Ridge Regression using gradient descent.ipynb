```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=17)
```

```python
# we use SGDRegressor
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
# penalty means adding regularization (l2='Ridge regularization')
```

```python
reg.fit(x_train,y_train)

y_pred=reg.predict(x_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
# We use Ridge
```

```python
from sklearn.linear_model import Ridge
r=Ridge(alpha=0.001,max_iter=100,solver='sparse_cg')
```

```python
r.fit(x_train,y_train)

y_pred1=r.predict(x_test)
print("R2_score",r2_score(y_test,y_pred1))
print(r.coef_)
print(r.intercept_)
```

```python
class MeraRidgeGD:
    def __init__(self,learning_rate,alpha,epochs):
        self.learning_rate=learning_rate
        self.alpha=alpha
        self.epochs=epochs
        self.coef_=None
        self.intercept_=None
        
    def fit(self,x_train,y_train):
        self.coef_=np.ones(x_train.shape[1])
        self.intercept_=0
        theta=np.insert(self.coef_,0,self.intercept_)
        
        x_train=np.insert(x_train,0,1,axis=1)
        
        for i in range(self.epochs):
            theta_der=np.dot(x_train.T,x_train).dot(theta)-np.dot(x_train.T,y_train)+self.alpha*theta
            theta=theta-self.learning_rate*theta_der
        self.coef_=theta[1:]
        self.intercept_=theta[0]
        
    def predict(self,x_test):
        return np.dot(x_test,self.coef_)+self.intercept_
        
```

```python
regression=MeraRidgeGD(epochs=500,alpha=0.001,learning_rate=0.005)
```

```python
regression.fit(x_train,y_train)

y_pred2=regression.predict(x_test)
print("R2_score",r2_score(y_test,y_pred2))
print(regression.coef_)
print(regression.intercept_)
```
