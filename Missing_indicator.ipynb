```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer,MissingIndicator
```

```python
df=pd.read_csv("C:/Users/sanke/Downloads/train (1).csv",usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
x=df.drop(columns='Survived')
y=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
x_train.head()
```

# Without using Missing indictor

```python
si=SimpleImputer()
x_train_trf=si.fit_transform(x_train)
x_test_trf=si.transform(x_test)
```

```python
x_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(x_train_trf,y_train)

y_pred = clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Using MissingIndicator

```python
mi = MissingIndicator()
mi.fit(x_train)
```

```python
mi.features_
```

```python
x_train_missing=mi.transform(x_train)
```

```python
x_train_missing
```

```python
x_test_missing=mi.transform(x_test)
```

```python
x_test_missing
```

```python
x_train['Age_NA']=x_train_missing
```

```python
x_train
```

```python
x_test['Age_NA']=x_test_missing
```

```python
x_test
```

```python
si = SimpleImputer()

X_train_trf2 = si.fit_transform(x_train)
X_test_trf2 = si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred2 = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred2)
```

# Using Simple Imputer's Class

```python
si=SimpleImputer(add_indicator=True)
```

```python
X_train2=si.fit_transform(x_train)
```

```python
X_test2=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train2,y_train)
y_pred=clf.predict(X_test2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
